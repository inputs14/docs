# Two Stack - Java
* Linear data structure.
* Implement in an order LIFO - Last in First Out or FILO - Fast In Last Out.
* Stacks should have bounded capacity.

## Applications
* Balancing of symbols. Eg: [()()]{[()]}
* Reversing Eg: welcome => emoclew
* Backtracing Eg: During debugging of an application.
* Undo mechanism 
* Redo mechanism(With Two Stacks)

```java
public class TwoStacks {
    private int[] items;
    private int top1;
    private int top2;
    
    public TwoStacks(int length){
        items = new int[length];
        top1 = 0;
        top2 = length;
    }
    
    public boolean isFull(){
        if((top1 - top2) == 0)
            return true;
        
        return false;
    }
    
    public boolean isEmpty1(){
        // For Stack 1
        if(top1 == 0)
            return true;
        
        return false;
    }
    
    public boolean isEmpty2(){
        // For Stack 2
        if(top2 == items.length)
            return true;
        
        return false;
    }
    
    public void push1(int value){
        //For stack 1
        if(isFull()){
            System.out.println("Stacks are Full");
            return;
        }
            
        items[top1] = value;
        top1++;
    }
    
    public void push2(int value){
        //For stack 2
        if(isFull()){
            System.out.println("Stacks are Full");
            return;
        }
            
        items[top2-1] = value;
        top2--;
    }
    
    public int pop1(){
        // For Stack 1
        if(isEmpty1()){
            System.out.println("Stack 1 is Empty");
            return -1;
        }
            
        
        int return_val = items[top1-1];
        top1--;
        return return_val;
    }
    
    public int pop2(){
        // For Stack 2
        if(isEmpty2()){
            System.out.println("Stack 2 is Empty");
            return -1;
        }
            
        
        int return_val = items[top2+1];
        top2++;
        return return_val;
    }
    
    public int peek1(){
        // For Stack 1
        if(isEmpty1()){
            System.out.println("Stack 1 is Empty");
            return -1;
        }
            
        int return_val = items[top1-1];
        return return_val;
    }
    
    public int peek2(){
        // For Stack 2
        if(isEmpty2()){
            System.out.println("Stack 1 is Empty");
            return -1;
        }
            
        
        int return_val = items[top2+1];
        return return_val;
    }
}
```

```java
TwoStacks stack = new TwoStacks(7);
stack.push1(1);
stack.push2(-1);
stack.push1(2);
stack.push2(-2);
stack.push2(-3);
stack.push2(-4);
System.out.println("Stack 1 Peek - "+stack.peek1());
System.out.println("Stack 2 Peek - "+stack.peek2());

stack.push1(3);
stack.push2(-5);

System.out.println("Stack 1 Popped out - "+stack.pop1());
System.out.println("Stack 2 Popped out - "+stack.pop2());
System.out.println("Stack 1 Popped out - "+stack.pop1());

System.out.println("Stack 1 Peek - "+stack.peek1());
System.out.println("Stack 2 Peek - "+stack.peek2());
```
